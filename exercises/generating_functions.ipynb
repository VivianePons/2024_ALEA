## Binary words

Words in the alphabet $\{0,1 \}$

Coefficients should be $1,2,4,8,16,....$

## Palindromes

Palindromes in the alphabet $\{a, b \}$, i.e. words that read the same backward as forward.

Coefficients should be $1,2,2,4,4,8,8,....$

## Fibonnaci words

A word on the alphabet $\{a, b \}$ that never contains the factor $bb$.

Coefficients should be $1,2,3,5,8,13,21,...$

## Ternary trees

A ternary tree is either a leaf or a node with $3$ children which are also ternary trees. The size of the tree is the number of internal nodes.

Coefficients should be $1,1,3,12,55,273,1428,....$

## Trees with binary / ternary levels

Trees where the root has two children, level 1 nodes (children of the root) have 3 children, level 2 nodes have 2 children, etc.
Children can all also be leaves (0 children). The size of the tree is the number of internal nodes.

Coefficients should be $1,1,2,7,24,95,386,...$

### A binary tree of fibonnaci words

A binary tree in which each internal node cotains a non-empty fibonacci word. The size of the tree is the sum of the sizes of the words.

*Note: look at possible methods on a defined lazy series*

Coefficients should be $1,2,11,69,470,3395, 25578$

## A binary tree of binary trees

A binary tree in which each internal node contains another (non empty) binary tree. The size is the sum of sizes of all internal binary trees.

Coefficients should be $1,1,4,18,86,427,2180$
